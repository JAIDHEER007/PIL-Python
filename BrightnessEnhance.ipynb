```python
from PIL import Image
from PIL import ImageEnhance
import urllib.request
from IPython.display import display
```

```python
# Retreiving the image from Internet
url = 'https://vetstreet.brightspotcdn.com/dims4/default/4f4dea1/2147483647/crop/0x0%2B0%2B0/resize/645x380/quality/90/?url=https%3A%2F%2Fvetstreet-brightspot.s3.amazonaws.com%2F31%2F1ba400a28511e087a80050568d634f%2Ffile%2FBengal-3-645mk062211.jpg'
urllib.request.urlretrieve(url, 'cat.jpg')
```

```python
img1 = Image.open('cat.jpg')
```

```python
display(img1)
```

```python
print(img1.width)
print(img1.height)
print(img1.mode)
```

```python
# Resizing the image to half its size
img1 = img1.resize((322, 190))
```

```python
display(img1)
```

```python
# enhancing the brightness
# using ImageEnhance

enhancer = ImageEnhance.Brightness(img1)

# Enhanced Image
display(enhancer.enhance(0.0))  # Full Black, 0 brightness
display(enhancer.enhance(0.25))  
display(enhancer.enhance(0.5))  
display(enhancer.enhance(0.75))  
display(enhancer.enhance(1.0))  # Original Image
display(enhancer.enhance(1.5))  
display(enhancer.enhance(2.0))  
```
