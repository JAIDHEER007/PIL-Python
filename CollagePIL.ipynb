```python
from PIL import Image
import urllib.request
from IPython.display import display
```

```python
# Retreiving the image from Internet
url = 'https://vetstreet.brightspotcdn.com/dims4/default/4f4dea1/2147483647/crop/0x0%2B0%2B0/resize/645x380/quality/90/?url=https%3A%2F%2Fvetstreet-brightspot.s3.amazonaws.com%2F31%2F1ba400a28511e087a80050568d634f%2Ffile%2FBengal-3-645mk062211.jpg'
urllib.request.urlretrieve(url, 'cat.jpg')
```

```python
img = Image.open('cat.jpg')
```

```python
img.mode
```

```python
img = img.convert('RGB')
```

```python
# Cat1
display(img)
```

```python
print(img.width)
print(img.height)
```

```python
# Resizing the image to 322 X 190 pixels (half the size of original)
img = img.resize((322, 190))
```

```python
display(img)
```

```python
# Creating a new image of size 2 times the previous one with Image.new
img2 = Image.new(mode = img.mode, size = (2 * img.width, 2 * img.height))
# Creates a black image 
```

```python
display(img2)
```

```python
# creating a collage of 4 cats
# using Image.paste and some math

x = 0
y = 0
for _ in range(4):
    # Pasting the image
    img2.paste(im = img, box = (x, y))

    if (x + img.width) == img2.width:
        x = 0
        y += img.height
    else: x += img.width
```
